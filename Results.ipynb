# d vs d1 analysis

```python
import torch  # type: ignore
import pickle
from helpers import parameter_scan
from visualization import plot_heatmap_fixed
# Fixed hyperparameters
n = m = 1000
num_epochs = 30
# reps = 1
s = 5.0
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Values to scan
d_values = [1, 2,3, 4,5, 6,7, 8]
d1 = [2, 4, None]
p = 0.5
lr = 1e-3 # 0.001
wd = 1e-5
K= 1
reps = 1
# Run parameter scan
results = parameter_scan(
    n=n, m=m, d=d_values, p=p, lr=lr, weight_decay=wd,
    num_epochs=num_epochs, reps=reps, s=s, device=device, d1=d1, K=K
)

# save results
with open("Data_2/scan_d_d1_2.pkl", "wb") as f:
    pickle.dump(results, f)
print("Results saved successfully!")
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem
from collections import defaultdict

# Load results
with open("Data_2/scan_d_d1_2.pkl", "rb") as f:
    results = pickle.load(f)

# Organize metrics by d1
metrics_by_d1 = defaultdict(lambda: {
    'd': [],
    'gt_acc': [],
    'gt_acc_err': [],
    'val_loss': [],
    'val_loss_err': [],
    'test_acc': [],
    'test_acc_err': [],
    'reconst': [],
    'reconst_err': []
})

for exp in results:
    d = exp['params']['d']
    d1 = exp['params'].get('d1', None)  # fallback to None if not present

    # Extract results
    gt_acc_list = exp['results']['gt_accuracy']
    val_loss_list = exp['results']['val_losses']
    test_acc_list = exp['results']['accuracy']
    reconst_list = exp['results']['reconstruction_errors']
    final_val_losses = [losses[-1] for losses in val_loss_list]

    # Append values to the correct group
    group = metrics_by_d1[d1]
    group['d'].append(d)
    group['gt_acc'].append(np.mean(gt_acc_list))
    group['gt_acc_err'].append(sem(gt_acc_list))
    group['val_loss'].append(np.mean(final_val_losses))
    group['val_loss_err'].append(sem(final_val_losses))
    group['test_acc'].append(np.mean(test_acc_list))
    group['test_acc_err'].append(sem(test_acc_list))
    group['reconst'].append(np.mean(reconst_list))
    group['reconst_err'].append(sem(reconst_list))

# Sort values within each group by d
for group in metrics_by_d1.values():
    sorted_data = sorted(zip(
        group['d'], group['gt_acc'], group['gt_acc_err'],
        group['val_loss'], group['val_loss_err'],
        group['test_acc'], group['test_acc_err'],
        group['reconst'], group['reconst_err']
    ))
    (group['d'], group['gt_acc'], group['gt_acc_err'],
     group['val_loss'], group['val_loss_err'],
     group['test_acc'], group['test_acc_err'],
     group['reconst'], group['reconst_err']) = map(list, zip(*sorted_data))

# === PLOTS ===

# Colors for each d1
color_map = {
    None: "blue",
    4: "green",
    2: "orange"
}

# 1. Accuracy vs d
plt.figure(figsize=(8, 5))
for d1, group in metrics_by_d1.items():
    label = f"d1 = {d1}" if d1 is not None else "d1 = None"
    plt.errorbar(group['d'], group['gt_acc'], yerr=group['gt_acc_err'],
                 fmt='o-', capsize=5, label=f"GT Accuracy ({label})", color=color_map[d1])
    plt.errorbar(group['d'], group['test_acc'], yerr=group['test_acc_err'],
                 fmt='s--', capsize=5, label=f"Validation Accuracy ({label})", color=color_map[d1])

plt.title("Accuracy vs d for different d1")
plt.xlabel("d (embedding dimension)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Results_2/combined_accuracy_vs_d.png")
plt.show()

# 2. Final Validation Loss vs d
plt.figure(figsize=(8, 5))
for d1, group in metrics_by_d1.items():
    label = f"d1 = {d1}" if d1 is not None else "d1 = None"
    plt.errorbar(group['d'], group['val_loss'], yerr=group['val_loss_err'],
                 fmt='o-', capsize=5, label=label, color=color_map[d1])

plt.title("Final Validation Loss vs d")
plt.xlabel("d (embedding dimension)")
plt.ylabel("Final Val Loss")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Results_2/final_val_loss_vs_d.png")
plt.show()

# 3. Reconstruction Error vs d
plt.figure(figsize=(8, 5))
for d1, group in metrics_by_d1.items():
    label = f"d1 = {d1}" if d1 is not None else "d1 = None"
    plt.errorbar(group['d'], group['reconst'], yerr=group['reconst_err'],
                 fmt='o-', capsize=5, label=label, color=color_map[d1])

plt.title("Reconstruction Error vs d")
plt.xlabel("d (embedding dimension)")
plt.ylabel("Reconstruction Error (Normalized)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Results_2/reconstruction_error_vs_d.png")
plt.show()

```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import sem

# Load the saved results
file_path_1 = "Data_2/s_scan_2.pkl"

with open(file_path_1, "rb") as f:
    results = pickle.load(f)

# with open(file_path_2, "rb") as f:
#     results.extend(pickle.load(f))  # ✅ Merge both datasets

# Lists for metrics
s_values = []
acc_means = []
acc_errors = []
loss_means = []
loss_errors = []
test_acc_means = []
test_acc_errors = []
reconst_means = []
reconst_errors = []

# Extract all relevant metrics per experiment
for exp in results:
    s = exp['params']['s']
    gt_acc_list = exp['results']['gt_accuracy']
    val_loss_list = exp['results']['val_losses']
    test_acc_list = exp['results']['accuracy']
    reconst_list = exp['results']['reconstruction_errors']

    final_val_losses = [losses[-1] for losses in val_loss_list]

    s_values.append(s)
    acc_means.append(np.mean(gt_acc_list))
    acc_errors.append(sem(gt_acc_list))
    loss_means.append(np.mean(final_val_losses))
    loss_errors.append(sem(final_val_losses))
    test_acc_means.append(np.mean(test_acc_list))
    test_acc_errors.append(sem(test_acc_list))
    reconst_means.append(np.mean(reconst_list))
    reconst_errors.append(sem(reconst_list))

# Sort by s
(s_sorted, acc_means, acc_errors, 
 loss_means, loss_errors, 
 test_acc_means, test_acc_errors, 
 reconst_means, reconst_errors) = zip(*sorted(
    zip(s_values, acc_means, acc_errors, loss_means, loss_errors, 
        test_acc_means, test_acc_errors, reconst_means, reconst_errors)
))

# === PLOTS ===

# 1. Accuracy vs s

plt.figure(figsize=(8, 5))
plt.errorbar(s_sorted, acc_means, yerr=acc_errors, fmt='o-', capsize=5, label='Ground Truth Accuracy')
plt.errorbar(s_sorted, test_acc_means, yerr=test_acc_errors, fmt='o-', capsize=5, label='Validation Accuracy', color='green')

plt.title("Accuracy vs s")
plt.xlabel("s (scale)")
plt.ylabel("Accuracy")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig("Results_2/combined_accuracy_vs_s.png")
plt.show()


# 2. Final Validation Loss vs s
plt.figure(figsize=(8, 5))
plt.errorbar(s_sorted, loss_means, yerr=loss_errors, fmt='o-', capsize=5, color='orange')
plt.title("Final Validation Loss vs s")
plt.xlabel("s (scale)")
plt.ylabel("Final Val Loss")
plt.grid(True)
plt.tight_layout()
plt.savefig("Results_2/final_val_loss_vs_s.png")
plt.show()



# 3. Reconstruction Error vs s
plt.figure(figsize=(8, 5))
plt.errorbar(s_sorted, reconst_means, yerr=reconst_errors, fmt='o-', capsize=5, color='red')
plt.title("Reconstruction Error vs s")
plt.xlabel("s (scale)")
plt.ylabel("Reconstruction Error (Normalized)")
plt.grid(True)
plt.tight_layout()
plt.savefig("Results_2/reconstruction_error_vs_s.png")
plt.show()
```
